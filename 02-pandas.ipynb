```python
import pandas as pd

dataset = pd.read_csv("data/dataset.csv")
# pd.read_csv("data/dataset.csv", index_col="id")

# dataset.head() # muestra las primeas 5 

# dataset.describe()

dataset
```

## Limpieza de datos

```python
# cuando hay datos vacios podemos eliminar esos datos
dataset_filtrado = dataset.dropna() # elimina las filas completas < -- > 
dataset_filtrado

# cambiar el valor de NAN
dataset_filtrado = dataset.fillna(0) # cambiar valor NAN por 0

# limpiar multiples datos
dataset_filtrado = dataset.fillna({'retweets': 0, 'mentions': 0})
```

## Filtrado de columnas

```python
# mostrar sola una columna full_text / los datos se repiten
dataset_filtrado['full_text']

# dataset_filtrado['country']

# mostrar 2 columnas
dataset_filtrado[['full_text', 'country']]
```

## Filtrado por filas

```python
# mostrar las 10 primeras filas
dataset.iloc[0:10]

# mostrar de 10 - 20 filas
dataset.iloc[10:20]

# mostrar de 10 - 20 filas con filtrado de texto y country
dataset.loc[10: 20, ['full_text', 'country']]
```

## Filtrado por condicion

```python
# mostrar los country de ecuador
# dataset_filtrado[dataset_filtrado['country'] == "ECUADOR"]

# mostrar los followers mayor a 300
# dataset_filtrado[dataset_filtrado['followers'] > 300]

# mostrar los followers menor a 300
dataset_filtrado[dataset_filtrado['followers'] < 300]

# 2 condiciones mostrar tweets favoritos arriba de 400 y con followers mayor a 900
dataset_filtrado[(dataset_filtrado['favorites'] > 400) & (dataset_filtrado['followers'] > 900)]

# mostra datos con reetweets mayor o igual a 123 y country mexico
dataset_filtrado[(dataset_filtrado['retweets'] >= 123) & (dataset_filtrado['country'] == "MEXICO")]

# filtrar por texto Programming
dataset_filtrado[dataset_filtrado['full_text'].str.contains('Programming')]

```

## Crear nuevos datos


```python
dataset_filtrado

def followersAndFollowees(row):
    return row['followers'] + row['followees']

# Aplica la funci√≥n a cada fila del DataFrame
dataset_filtrado['followersAndFollowees'] = dataset_filtrado.apply(followersAndFollowees, axis=1)

dataset_filtrado
```
